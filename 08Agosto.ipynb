```python
# AGOSTO - DINAC
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data_path16 = './2016-20230315T011102Z-001/2016/2016_08'
data_path17 = './2017-20230315T011107Z-001/2017/2017_08'
data_path18 = './2018-20230315T011112Z-001/2018/2018_08'
mes = 'Agosto'
anho16 = '2016'
anho17 = '2017' 
anho18 = '2018'
```

```python
files16 = [f for f in os.listdir(data_path16) if f.startswith('Paraguay')]
files17 = [f for f in os.listdir(data_path17) if f.startswith('Paraguay')]
files18 = [f for f in os.listdir(data_path18) if f.startswith('Paraguay')]
```

```python
df16 = None
df17 = None
df18 = None
m16 = len(files16)
m17 = len(files17)
m18 = len(files18)
```

```python
# Se a√±ade una columna que indique el dia del mes y se crea el dataframe
for k in range (0,m16):
    if os.stat(data_path16+'/'+files16[k]).st_size != 0 :
        aux = pd.read_csv(data_path16+'/'+files16[k])
        aux['day'] = int(files16[k][20:22])
        df16 = pd.concat([df16,aux], ignore_index = True)
for k in range (0,m17):
    if os.stat(data_path17+'/'+files17[k]).st_size != 0 :
        aux = pd.read_csv(data_path17+'/'+files17[k])
        aux['day'] = int(files17[k][20:22])
        df17 = pd.concat([df17,aux], ignore_index = True)
for k in range (0,m18):
    if os.stat(data_path18+'/'+files18[k]).st_size != 0 :
        aux = pd.read_csv(data_path18+'/'+files18[k])
        aux['day'] = int(files18[k][25:27])
        df18 = pd.concat([df18,aux], ignore_index = True)
```

```python
#Tipos de Relampagos
#2016
df16[['day','type']].plot.scatter(x='day',y='type')
plt.xlabel('Dia')
plt.ylabel('Tipo de relampago')
plt.title('Tipos de relampago '+mes+'_'+anho16+' DINAC')
#2017
df17[['day','type']].plot.scatter(x='day',y='type')
plt.xlabel('Dia')
plt.ylabel('Tipo de relampago')
plt.title('Tipos de relampago '+mes+'_'+anho17+' DINAC')
#2018
df18[['day','type']].plot.scatter(x='day',y='type')
plt.xlabel('Dia')
plt.ylabel('Tipo de relampago')
plt.title('Tipos de relampago '+mes+'_'+anho18+' DINAC')
```

```python
## Tipos de relampagos representados porcentualmente
#2016
labels = ['Entre Nubes','Nube a Tierra','Nube a Tierra Global']
size = [int(df16['type'].value_counts()[1]),int(df16['type'].value_counts()[0]),int(df16['type'].value_counts()[40])]
colors = ['gray','orange','cyan']
explode = [0,0.1,0]
plt.pie(size, labels = labels, colors = colors,explode=explode, shadow = True, autopct = '%.2f%%')
plt.title('Tipos de Rayos '+mes+'_'+anho16+' DINAC', fontsize  = 10)
plt.axis('off')
plt.legend(loc= 'best')
plt.show()
#2017
labels = ['Entre Nubes','Nube a Tierra','Nube a Tierra Global']
size = [int(df17['type'].value_counts()[1]),int(df17['type'].value_counts()[0]),int(df17['type'].value_counts()[40])]
colors = ['gray','orange','cyan']
explode = [0,0.1,0]
plt.pie(size, labels = labels, colors = colors,explode=explode, shadow = True, autopct = '%.2f%%')
plt.title('Tipos de Rayos '+mes+'_'+anho17+' DINAC', fontsize  = 10)
plt.axis('off')
plt.legend(loc= 'best')
plt.show()
#2018
labels = ['Entre Nubes','Nube a Tierra']
size = [int(df18['type'].value_counts()[1]),int(df18['type'].value_counts()[0])]
colors = ['gray','orange']
explode = [0,0.1]
plt.pie(size, labels = labels, colors = colors,explode=explode, shadow = True, autopct = '%.2f%%')
plt.title('Tipos de Rayos '+mes+'_'+anho18+' DINAC', fontsize  = 10)
plt.axis('off')
plt.legend(loc= 'best')
plt.show()
```
